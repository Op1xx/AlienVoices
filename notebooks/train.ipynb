# AlienVoices — Обучение модели
Подготовка данных и обучение классификатора голосов.

```python
import sys
sys.path.insert(0, '..')

import os
import numpy as np
from app.ml.preprocessor import extract_features
from app.ml.model import build_model, train_model
from config import DATA_RAW_PATH, NUM_CLASSES
```

## 1. Загрузка данных

```python
# Ожидаемая структура: data/raw/<class_id>/<file.wav>
X, y = [], []

for class_id in range(NUM_CLASSES):
    class_dir = os.path.join(DATA_RAW_PATH, str(class_id))
    if not os.path.exists(class_dir):
        continue
    for fname in os.listdir(class_dir):
        if not fname.endswith(('.wav', '.mp3', '.flac')):
            continue
        features = extract_features(os.path.join(class_dir, fname))
        X.append(features)
        y.append(class_id)

X = np.array(X)
y = np.array(y)
print(f'Загружено: {len(X)} образцов, классов: {NUM_CLASSES}')
```

## 2. Обучение

```python
model = train_model(X, y)
print('Модель сохранена.')
```
